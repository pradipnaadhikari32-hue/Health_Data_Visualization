```python
# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Health_heart_experimental.csv")

# Check first 5 rows
df.head()
```

```python
df.info()
df.describe()
```

```python
# 1. Line Chart â€“ Average Heart Rate by Age
avg_hr_age = df.groupby('age')['HR'].mean()

plt.figure()
plt.plot(avg_hr_age.index, avg_hr_age.values)
plt.xlabel('Age')
plt.ylabel('Average Heart Rate')
plt.title('Average Heart Rate by Age')
plt.show()
```

1.Line Chart â€“ Average Heart Rate by Age

Purpose:
To analyze how heart rate changes with age.

Finding:
The average heart rate remains relatively stable across different age groups with minor fluctuations. This suggests that age alone does not significantly affect heart rate without other health conditions.

```python
# 2. Bar Plot â€“ Sex Distribution
plt.figure()
df['sex'].value_counts().plot(kind='bar')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.title('Sex Distribution')
plt.show()
```

2. Bar Plot â€“ Sex Distribution

Purpose:
To compare the number of male and female patients.

Finding:
The dataset contains almost equal numbers of males and females, which means the data is balanced and suitable for fair analysis.

```python
# 3. Pie Chart â€“ Heart Disease Indication
plt.figure()
df['indication'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.title('Heart Disease Indication Distribution')
plt.show()
```

3.Pie Chart â€“ Heart Disease Indication

Purpose:
To show the percentage of people with and without heart disease.

Finding:

Around 52.7% of individuals do not have heart disease.

Around 47.3% do have heart disease.
This indicates a significant presence of heart-related issues in the dataset.

```python
# 4. Box Plot â€“ BMI Distribution
plt.figure()
plt.boxplot(df['BMI'])
plt.ylabel('BMI')
plt.title('BMI Distribution')
plt.show()
```

4. Box Plot â€“ BMI Distribution

Purpose:
To identify BMI spread and detect outliers.

Finding:
Most BMI values lie between 24 and 35, with a median around 29, indicating many individuals are overweight, which is a known risk factor for heart disease.

```python
# 5. Histogram â€“ Age Distribution
plt.figure()
plt.hist(df['age'], bins=20)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

5. Histogram â€“ Age Distribution

Purpose:
To observe how age is distributed in the dataset.

Finding:
Ages are evenly distributed between 20 and 70 years, showing that the dataset includes both young and older adults.

```python
# 6. Scatter Plot â€“ BMI vs Systolic Blood Pressure
plt.figure()
plt.scatter(df['BMI'], df['SysBP'])
plt.xlabel('BMI')
plt.ylabel('Systolic Blood Pressure')
plt.title('BMI vs Systolic Blood Pressure')
plt.show()
```

Scatter Plot â€“ BMI vs Systolic Blood Pressure

Purpose:
To analyze the relationship between BMI and systolic blood pressure.

Finding:
There is a visible positive trendâ€”higher BMI values tend to correspond with higher systolic blood pressure, indicating increased cardiovascular risk.

Conclusion

The visualizations help understand patterns in heart health data. Factors such as BMI, blood pressure, and age play an important role in heart disease risk. Data visualization makes it easier to identify these trends and supports better health analysis and decision-making.
